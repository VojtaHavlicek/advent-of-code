```python
# ------ LOAD ARRAY ----- 
array = []

x = 0
y = 0

directions = [(0, -1), (1, 0), (0, 1), (-1, 0)]
direction = 0

with open("input.txt") as file: 
    for line in file: 
        array.append(list(line.strip()))

        if '^' in line: 
            x = line.index('^')
            y = len(array) - 1

["".join(line) for line in array]
print(x,y)

# ------- TRAVERSE --------
loop_counter = 0 

while True:

    dirx, diry = directions[direction]

    # Check if we left the map 
    if y+diry < 0 or y+diry >= len(array) or x+dirx < 0 or x+dirx >= len(array[0]):
        break

    # Turn if you are about to hit the wall
    if array[y+diry][x+dirx] == '#': 
        direction = (direction + 1) % 4
        continue    


    # Place an obstacle in front of you if you can 
    if array[y+diry][x+dirx] != '^' and array[y+diry][x+dirx] != 'X' : # Cannot place the obstacle at the starting point or a point that has already been visited

        # Simulate right turn after placing an obstruction
        temp_direction = (direction + 1) % 4
    
        temp_x = x 
        temp_y = y 
        temp_path = set((x,y, temp_direction))
        
        # Simulate the path
        while True: 
            temp_dirx, temp_diry = directions[temp_direction]
            # Check if we left the map
            if temp_y + temp_diry < 0 or temp_y + temp_diry >= len(array) or temp_x + temp_dirx < 0 or temp_x + temp_dirx >= len(array[0]):
                break


            # Turn to the right if you hit a wall. Remember that you placed the new obstacle! 
            # NOTE: WTF. I did not test for the starting point here (:( ) 
            if array[temp_y + temp_diry][temp_x + temp_dirx] == '#' or (temp_y + temp_diry == y + diry and temp_x + temp_dirx == x + dirx):
                temp_direction = (temp_direction + 1) % 4
                temp_path.add((temp_x, temp_y, temp_direction)) 
                continue
            
            # Take a step
            temp_x += temp_dirx
            temp_y += temp_diry


            # If you hit the starting point, check if you have a loop
            if (temp_x, temp_y, temp_direction) in temp_path:
                loop_counter += 1  
                break

             # Add the current position to the path, including the direction
            temp_path.add((temp_x, temp_y, temp_direction))  

    # Move 
    x += dirx
    y += diry

    # Mark the path
    array[y][x] = 'X'

loop_counter
```
