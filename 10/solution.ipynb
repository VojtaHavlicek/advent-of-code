```python
lines = []

with open("input.txt") as f:
    for line in f:
        lines.append(list(map(int,list(line.strip()))))
        
print(lines)

trailheads = []

for row in range(len(lines)):
    for col in range(len(lines[row])):
        if lines[row][col] == 0:
            trailheads.append((row, col))
        



trailheads
```

```python
total_score = 0
for trailhead in trailheads: 
    current = [trailhead]
    visited = set()

    score = 0
    while len(current) > 0:
        row, col = current.pop(0)

        height = lines[row][col]
       
        if height == 9: # and (row, col) not in visited:
            score += 1
            #visited.add((row, col))
            continue

        if row > 0 and lines[row-1][col] == height + 1:
            current.append((row-1, col))

        if row < len(lines) - 1 and lines[row+1][col] == height + 1:
            current.append((row+1, col))

        if col > 0 and lines[row][col-1] == height + 1:
            current.append((row, col-1))

        if col < len(lines[0]) - 1 and lines[row][col+1] == height + 1:
            current.append((row, col+1))

    total_score += score

total_score
    
```
