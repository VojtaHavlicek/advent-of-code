```python
array = []
with open("input.txt") as f:
    for line in f:
        array.append(list(line.strip()))

fence_data = dict() 
temp_array = deepcopy(array)

area_counter = 0

for row in range(len(array)):
    for col in range(len(array[row])):
        if temp_array[row][col] == '.':
            continue

        type = array[row][col]

      
        # find the region 
        visited = []
        perimeter_tiles = set()
        to_visit = [(row, col)]

        perimeter = 0 
        sides = 0
        
        while len(to_visit) > 0: 
            temp_row, temp_col = to_visit.pop()
            if (temp_row, temp_col) in visited:
                continue

            if array[temp_row][temp_col] == type:
                visited.append((temp_row, temp_col))
                temp_array[temp_row][temp_col] = '.'

                if temp_row-1 < 0 or array[temp_row - 1][temp_col] != type:
                    perimeter += 1
                    perimeter_tiles.add((temp_row, temp_col))
                if temp_row+1 >= len(temp_array) or array[temp_row + 1][temp_col] != type:
                    perimeter += 1
                    perimeter_tiles.add((temp_row, temp_col))
                if temp_col-1 < 0 or array[temp_row][temp_col - 1] != type:
                    perimeter += 1
                    perimeter_tiles.add((temp_row, temp_col))
                if temp_col+1 >= len(temp_array[0]) or array[temp_row][temp_col + 1] != type:
                    perimeter += 1
                    perimeter_tiles.add((temp_row, temp_col))
            else:
                continue

            neighbors = [
                (temp_row - 1, temp_col),
                (temp_row + 1, temp_col),
                (temp_row, temp_col - 1),
                (temp_row, temp_col + 1)
            ]
            neighbors = [(r, c) for r, c in neighbors if r >= 0 and r < len(array) and c >= 0 and c < len(array[r])]

            to_visit.extend(neighbors)
 
        # find the number of corners in the perimeter
        sides = 0
        area = len(visited)

        # UFF, this was not elegant. 
        for tile in visited:
            temp_row, temp_col = tile 

            edges = {"top":0, "bottom":0, "left":0, "right":0}

            if temp_row-1 < 0 or array[temp_row-1][temp_col] != type:
               edges["top"] = 1
            if temp_row+1 >= len(array) or array[temp_row+1][temp_col] != type:
               edges["bottom"] = 1
            if temp_col-1 < 0 or array[temp_row][temp_col-1] != type:
                edges["left"] = 1
            if temp_col+1 >= len(array[temp_row]) or array[temp_row][temp_col+1] != type:
                edges["right"] = 1

            # Outer corners
            if edges["top"] and edges["left"]:
                sides += 1
            
            if edges["top"] and edges["right"]:
                sides += 1

            if edges["bottom"] and edges["left"]:
                sides += 1
            
            if edges["bottom"] and edges["right"]:
                sides += 1

            # Inner corners? 
            if temp_row + 1 < len(array): 
                # @@
                # @.
                if temp_col + 1 < len(array[temp_row]):
                    if array[temp_row+1][temp_col] == type and array[temp_row][temp_col+1] == type and array[temp_row+1][temp_col+1] != type:
                        sides += 1

                        
                            
                if temp_col - 1 >= 0:
                    # @@
                    # .@
                    if array[temp_row+1][temp_col] == type and array[temp_row][temp_col-1] == type and array[temp_row+1][temp_col-1] != type:
                        sides += 1

                      

            if temp_row - 1 >= 0:
                # @.
                # @@
                if temp_col + 1 < len(array[temp_row]):
                    if array[temp_row-1][temp_col] == type and array[temp_row][temp_col+1] == type and array[temp_row-1][temp_col+1] != type:
                        sides += 1

                      

                if temp_col - 1 >= 0:
                    # .@
                    # @@
                    if array[temp_row-1][temp_col] == type and array[temp_row][temp_col-1] == type and array[temp_row-1][temp_col-1] != type:
                        sides += 1
        

        fence_data[type+'_'+str(area_counter)] = {"area":area, "perimeter":perimeter, "sides":sides}
        area_counter += 1

        # find the perimeter
        
temp_array
array
        
```
