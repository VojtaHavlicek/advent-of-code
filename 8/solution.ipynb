```python
array = []

antenas = dict() 

with open("input.txt") as f:
    for row, line in enumerate(f):
        row_content = list(line.strip())
        array.append(row_content)

        for col, entry in enumerate(row_content):
            if entry == ".":
                continue
            if entry not in antenas:
                antenas[entry] = [(row,col)]
            else:
                antenas[entry].append((row,col))


array
```

First, filter distinct frequencies and get coordinate pairs.
Then for every frequency, compute the antinodes within bounds.

```python


antipodes = 0 

for frequency, locations in antenas.items():
    # Iterate over pairs of locations and compute the location of antipodes 
    for (ix,iy) in locations:
        for (jx,jy) in locations:
            if (ix,iy) == (jx,jy): 
                continue

            # Compute the antipode of the pair
            k = 0
            while True:
                ax = ix - k*(jx - ix)
                ay = iy - k*(jy - iy)

                # Check if the antipode is within the array
                if ax < 0 or ax >= len(array) or ay < 0 or ay >= len(array[0]):
                    break
            
                array[ay][ax] = "#"

                k += 1


for row in array:
    for entry in row:
        if entry == "#":
            antipodes += 1

antipodes
```
