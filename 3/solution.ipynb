```python
with open("input.txt", "r") as f:
    lines = f.readlines()

contents = "".join(lines)
contents
```

```python
import re 
muls = re.findall("mul\(\d+,\d+\)", contents)
```

```python
result = 0 
for mul in muls: 
    a, b = map(int, mul[4:-1].split(","))
    result += a*b
```

* !!! CAREFUL WITH re.DOTALL !!!

```python
import re 

text = """do()This is the text 
between.don't()Another one.don't()"""
pattern = r"do\(\)(.*?)don't\(\)" # Match the entire block
enabled = re.findall(pattern, contents, re.DOTALL)
len(enabled) # 21 with DOTALL, 19 WITHOUT
```

```python
result = 0 
for line in enabled:
    muls = re.findall("mul\(\d+,\d+\)", line)
   
    print(muls)
    for mul in muls: 
        
        a, b = map(int, mul[4:-1].split(","))
        print(a,b)
        result += a*b

result
```
