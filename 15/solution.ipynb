```python
from copy import deepcopy

array = []
instructions = []
parse_instructions = False

with open("input.txt") as file:
    for line in file: 

        # Parse map:
        if not parse_instructions:
            if line == "\n":
                parse_instructions = True
                continue 

            line = list(line.strip())

            # Find start:
            if '@' in line:
                start = (line.index('@'), len(array))

            # Add another line 
            array.append(line)

        # Parse instructions:
        else: 
            instructions.append(line.strip())

instructions = list(''.join(instructions))

init_array = deepcopy(array)
```

```python
position = start
array = deepcopy(init_array)
array[start[1]][start[0]] = '.' 

# Sokoban! 
for instruction in instructions:

    dirx, diry = 0, 0

    # Propose a new position for the box
    match instruction:
        case '>':
            dirx = 1 
        case '<':
            dirx = -1
        case '^':
            diry = -1
        case 'v':
            diry = 1

    next_position = (position[0] + dirx, position[1] + diry)

    # Take a step if the next position is empty: 
    if array[next_position[1]][next_position[0]] == '.':
        position = next_position
        continue

    # Do nothing if you hit a wall: 
    if array[next_position[1]][next_position[0]] == '#':
        continue

    # Try to push the box if you can: 
    if array[next_position[1]][next_position[0]] == 'O':

        box_position = next_position
        boxes = [] 
        
        while True: 
            next_box_position = (box_position[0] + dirx, box_position[1] + diry)
            
            if array[next_box_position[1]][next_box_position[0]] == 'O':
                boxes.append(next_box_position)
                box_position = next_box_position
            else: 
                break 

        # Push the boxes if there is space 
        if array[next_box_position[1]][next_box_position[0]] == '.':
            position = next_position
            array[position[1]][position[0]] = '.'

            for box in boxes:
                array[box[1]][box[0]] = 'O'

            array[next_box_position[1]][next_box_position[0]] = 'O'
            continue # And push all the boxes

        # Pass 
        if array[next_box_position[1]][next_box_position[0]] == '#':
            continue # And do nothing 
```

```python
array = deepcopy(init_array)

wide_array = [] 

for line in array:
    wide_line = [] 

    for char in line:
        match char:
            case "#": 
                wide_line.extend(["#","#"])
            case "O":
                wide_line.extend(list("[]"))
            case ".":
                wide_line.extend(list(".."))
            case "@":
                wide_line.extend(list("@."))

    wide_array.append(wide_line)

["".join(line) for line in wide_array]
```

```python
def wide_cost(array):
    H = len(array)
    W = len(array[0])

    result = 0 

    for row in range(H):
        for col in range(W):
            if array[row][col] == '[':
               result += 100*row + col

    return result
```

```python



# Wide array sweep 
array = deepcopy(wide_array)



from PIL import Image
import numpy as np

def draw_array(input_array):
    W = len(input_array[0])
    H = len(input_array)
    array = np.zeros((W, H, 3), dtype=np.uint8)
    for y in range(H):
        for x in range(W):
            match input_array[y][x]:
                case '#':
                    array[x, y] = [0, 0, 0]
                case '.':
                    array[x, y] = [255, 255, 255]
                case '@':
                    array[x, y] = [255, 0, 0]
                case '[' | ']':
                    array[x, y] = [0, 0, 255]

    img = Image.fromarray(array)
    return img 


# Find start:
for row in range(len(array)):
    for col in range(len(array[0])):
        if array[row][col] == '@':
            start = (col, row)
            break 

position = start
array[start[1]][start[0]] = '.' 

# Sokoban, just wide! 
for k, instruction in enumerate(instructions):

    # Debugging
    print_array = deepcopy(array)
    print_array[position[1]][position[0]] = '@'
    draw_array(print_array).save(f"frames/frame_{k:03d}.png")

    #print("".join(["".join(line) + "\n" for line in print_array]))
    #print(position, instruction)


    dirx, diry = 0, 0

    # Propose a new position for the box
    match instruction:
        case '>':
            dirx = 1 
        case '<':
            dirx = -1
        case '^':
            diry = -1
        case 'v':
            diry = 1

    next_position = (position[0] + dirx, position[1] + diry)

    # Take a step if the next position is empty: 
    if array[next_position[1]][next_position[0]] == '.':
        position = next_position
        continue

    # Do nothing if you hit a wall: 
    if array[next_position[1]][next_position[0]] == '#':
        continue

    # Try to push the box if you can: 
    if array[next_position[1]][next_position[0]] in ['[',']'] :

        box_position = next_position
        boxes = [box_position]
        

        # Vertical push: 
        if diry != 0: 
            # Check what side of the box does the robot push 
            if array[box_position[1]][box_position[0]] == '[': 
                boxes.append((box_position[0]+1, box_position[1]))
            else:
                boxes.append((box_position[0]-1, box_position[1]))

            # -----------------------------
            # PROPAGATE THE VERTICAL PUSH
            # -----------------------------
            frontier_boxes = set(boxes)
            dont_move = False  # Flag this if you propagate to a wall 

            while len(frontier_boxes) > 0:
                new_frontier = set() 
                for box in frontier_boxes: 
                    half_box = array[box[1]][box[0]]

                    if half_box == '[':
                        if array[box[1]+diry][box[0]] == ']':
                            new_frontier.add((box[0], box[1]+diry))
                            new_frontier.add((box[0]-1, box[1]+diry))

                        elif array[box[1]+diry][box[0]] == '[':
                            new_frontier.add((box[0], box[1]+diry)) # maybe also add the other half? 

                        elif array[box[1]+diry][box[0]] == '#':
                            dont_move = True 
                            break

                    if half_box == ']':
                        if array[box[1]+diry][box[0]] == '[':
                            new_frontier.add((box[0], box[1]+diry))
                            new_frontier.add((box[0]+1, box[1]+diry))

                        elif array[box[1]+diry][box[0]] == ']':
                            new_frontier.add((box[0], box[1]+diry)) # maybe also add the other half? 
                        
                        elif array[box[1]+diry][box[0]] == '#':
                            dont_move = True 
                            break
                        
                frontier_boxes = new_frontier
                boxes.extend(frontier_boxes)

            # Push the boxes if there is space
            if dont_move:
                continue
            
            # Check:
            assert(len(boxes) == len(set(boxes)))

            # Move all the boxes and the player
            for box in reversed(boxes):
                array[box[1]+diry][box[0]] = array[box[1]][box[0]]
                array[box[1]][box[0]] = '.'

            position = next_position # player moves
            array[position[1]][position[0]] = '.'
            continue

        # Horizontal push. Seems OK. 
        if dirx != 0:
            while True: 
                next_box_position = (box_position[0] + dirx, box_position[1])
                if array[next_box_position[1]][next_box_position[0]] in ['[',']']:
                    boxes.append(next_box_position)
                    box_position = next_box_position
                else: 
                    break 

            # Push the boxes if there is space 
            if array[next_box_position[1]][next_box_position[0]] == '.':
    
                for box in reversed(boxes):
                    array[box[1]][box[0]+dirx] = array[box[1]][box[0]]
                    array[box[1]][box[0]] = '.'

                position = next_position # player moves
                array[position[1]][position[0]] = '.'
                continue # And push all the boxes

        # Pass 
        if array[next_box_position[1]][next_box_position[0]] == '#':
            continue # And do nothing 


print_array = deepcopy(array)
print_array[position[1]][position[0]] = '@'



# print("".join(["".join(line) + "\n" for line in print_array]))
```
