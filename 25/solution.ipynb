```python
def get_heights(sequence):
    heights = [0 for _ in range(len(sequence[0]))]
    for col in range(len(sequence[0])): 
        height = 0 
        for row in range(len(sequence)):
            if sequence[row][col] == '#':
                height += 1
        heights[col] = height-1
    
    return heights
        


def get_keys_locks(): 
    keys = [] 
    locks = [] 

    keys_heights = [] 
    locks_heights = []

    with open("input.txt", "r") as file: 
        sequence = []
        for line in file:
            line = line.strip()
            if line:
                sequence.append(line)
            else:
                if '#' in sequence[-1]:
                    keys_heights.append(get_heights(sequence))
                    keys.append(sequence)
                else:
                    locks_heights.append(get_heights(sequence))
                    locks.append(sequence)
                sequence = []

        if '#' in sequence[-1]:
                    keys_heights.append(get_heights(sequence))
                    keys.append(sequence)
        else:
            locks_heights.append(get_heights(sequence))
            locks.append(sequence)
        sequence = []
    
    return keys, locks, keys_heights, locks_heights
```

```python
keys, locks, key_heights, locks_heights = get_keys_locks()
```

```python
result = 0 
height = len(keys[0])-2
# print(height)


for key in key_heights:
    for lock in locks_heights:
        print(key, lock)
        for i in range(len(key)):
            if lock[i] + key[i] > height:
                break
        else:
            result += 1

#print(result)
```
