```python
instructions = []

with open("input.txt") as f:
    
    for line in f:
        
        line = line.strip()
        if "Button A:" in line:  
            claw_machine = dict() 
            instruction_line = [entry.strip().split('+') for entry in line.split(":")[1].strip().split(",")]
            claw_machine["DAX"] = int(instruction_line[0][1])
            claw_machine["DAY"] = int(instruction_line[1][1])
        elif "Button B:" in line:
            instruction_line = [entry.strip().split('+') for entry in line.split(":")[1].strip().split(",")]
            claw_machine["DBX"] = int(instruction_line[0][1])
            claw_machine["DBY"] = int(instruction_line[1][1])
        elif "Prize:" in line:
            prize_line = [entry.strip().split('=') for entry in line.split(":")[1].strip().split(",")]
            claw_machine["X"]  = 10000000000000 + int(prize_line[0][1])
            claw_machine["Y"]  = 10000000000000 + int(prize_line[1][1])
            instructions.append(claw_machine)
```

```python
instructions
```

```python
score = 0
for instruction in instructions:
    denom = instruction["DAX"]*instruction["DBY"] - instruction["DBX"]*instruction["DAY"]
    a = (instruction["DBY"]*instruction["X"] - instruction["DBX"]*instruction["Y"])/denom 
    b = (instruction["DAX"]*instruction["Y"] - instruction["DAY"]*instruction["X"])/denom

    
    if a == int(a) and b == int(b): 
        print(f"Button A: {a}, Button B: {b}")
        score += 3*a + b
```
