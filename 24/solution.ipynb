```python
wire_values = {}
circuits = []

with open("input.txt", "r") as file:
    parse_wire_values = True 
    
    for line in file: 
        line = line.strip()
        if line == "":
            parse_wire_values = False
            continue

        if parse_wire_values: 
            line = line.split(':')
            wire_values[line[0]] = int(line[1].strip())

        else: 
            line = line.split(' -> ')
            output = line[1]
            inputs = line[0].split(' ')
            input1 = inputs[0]
            operator = inputs[1]
            input2 = inputs[2]
            circuits.append({"input1": input1, "operator": operator, "input2": input2, "output": output})
```

```python
from copy import deepcopy

def evaluate_circuit(circuits, wire_values):
    z_wires = {}

    _wire_values = deepcopy(wire_values)
    _circuits = deepcopy(circuits)

    while len(_circuits) > 0:
        circuit = _circuits.pop(0)
        input1 = circuit["input1"]
        operator = circuit["operator"]
        input2 = circuit["input2"]
        output = circuit["output"]

        if (input1 not in _wire_values) or (input2 not in _wire_values): 
            # move the circuit to the end of the list
            _circuits.append(circuit)
            continue

        match operator:
            case 'AND':
                _wire_values[output] = _wire_values[input1] & _wire_values[input2]
            case 'OR':
                _wire_values[output] = _wire_values[input1] | _wire_values[input2]
            case 'XOR': 
                _wire_values[output] = _wire_values[input1] ^ _wire_values[input2]

        if output[0] == 'z':
            z_wires[output] = _wire_values[output]

    import collections
    sorted_z_wires = collections.OrderedDict(sorted(z_wires.items()))
    sorted_z_wires

    answer = "".join(map(str, reversed(list(sorted_z_wires.values()))))

    try:
        z = int(answer, 2)
    except:
        print(f"Problem: {sorted_z_wires}, answer: {answer}")
    z = int(answer, 2)

    return z

# ----------------------------
print(f"x:{x}, y:{y}, z:{z}")
```

```python
# TODO: swap all circuits 
def get_swapped_circuits(circuits, swap_count = 2) -> list[dict] : 
    
    result = [] 
    wire_swaps = get_wire_swaps(len(circuits), swap_count)

    for wire_swap in wire_swaps:
        swapped_circuits = deepcopy(circuits)
        swap_sequence = []
        for i, j in wire_swap:
            swapped_circuits[i]['output'] = circuits[j]['output']
            swapped_circuits[j]['output'] = circuits[i]['output']

            swap_sequence.extend([circuits[i]['output'], circuits[j]['output']])

        result.append((sorted(swap_sequence), swapped_circuits))

    return result

```
