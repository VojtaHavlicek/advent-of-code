### Idea: 
Compose a tree out of the circuit list. 

```python
circuit_lines = []
output_to_inputs = {} 

with open("input.txt", "r") as file:
    parse_wire_values = True 
    
    for line in file: 
        line = line.strip()
        if line == "":
            parse_wire_values = False
            continue

        if parse_wire_values: 
            line = line.split(':')
            wire_values[line[0]] = int(line[1].strip())

            if line[0][0] == 'x':
                x_wires[line[0]] = int(line[1].strip())
            elif line[0][0] == 'y':
                y_wires[line[0]] = int(line[1].strip())

        else: 
            inputs, output = line.split(' -> ')
            circuit_lines.append((inputs, output))
            output_to_inputs[output] = inputs

circuit_lines
```

```python
from collections import OrderedDict

# Build syntatic tree

class Node:
    def __init__(self, operator=None, left=None, right=None, output=None):
        self.operator = operator    
        self.output = output
        self.left = left
        self.right = right

    def __repr__(self):
        if self.operator:
            return f'({self.left} {self.operator} {self.right})'
        return f'{self.output}'
    

def parse_expression(expression, output):
    # Split the expression into parts
    parts = expression.split()
    if len(parts) == 3:  # Format: 'A OPERATOR B'
        left, operator, right = parts
        return Node(operator=operator, left=left, right=right, output=output)
    else:
        raise ValueError(f"Invalid expression: {expression}")
    

def build_tree(rules):
    nodes = {}  # Dictionary to store nodes by their output name

    # First pass: Create nodes
    for expr, output in rules:
        nodes[output] = parse_expression(expr, output)

    # Second pass: Replace variable names with references to nodes
    for node in nodes.values():
        if node.left in nodes:
            node.left = nodes[node.left]
        if node.right in nodes:
            node.right = nodes[node.right]

    return nodes


def print_tree(node, indent=0):
    if isinstance(node, Node):
        print('  ' * indent + f'[{node.output}] {node.operator}')
        if isinstance(node.left, Node):
            print_tree(node.left, indent + 1)
        else:
            print('  ' * (indent + 1) + str(node.left))
        if isinstance(node.right, Node):
            print_tree(node.right, indent + 1)
        else:
            print('  ' * (indent + 1) + str(node.right))
    else:
        print('  ' * indent + str(node))


# Input rules
rules = circuit_lines

# Build the tree
nodes = build_tree(rules)

# Example: Print the tree for 'mjb'
print_tree(nodes)

# At this point, the terminal nodes have fully parsed expressions! 
nodes = OrderedDict(sorted(nodes.items()))
```

```python
def get_leaves(node): 
    result = [] 

    if isinstance(node, Node):
        if isinstance(node.left, Node):
            result.extend(get_leaves(node.left))
        else:
            result.extend([node.left])
        
        if isinstance(node.right, Node):
            result.extend(get_leaves(node.right))
        else:
            result.extend([node.right])

    else: 
        result.append(node)
    

    return set(result)
```

```python
def filter_subdictionary(nodes, keys):
    return {k: v for k, v in nodes.items() if k in keys}

def evaluate(node, wire_values):
    """ Evaluate the expression tree """

    if not isinstance(node, Node):
        return wire_values[node] # Leaf node

    left_leaves = get_leaves(node.left)
    right_leaves = get_leaves(node.right)

    left_leaves = filter_subdictionary(wire_values, left_leaves)
    right_leaves = filter_subdictionary(wire_values, right_leaves)
    
    if node.operator == 'AND':
        return evaluate(node.left, left_leaves) & evaluate(node.right, right_leaves)
    elif node.operator == 'OR':
        return evaluate(node.left, left_leaves) | evaluate(node.right, right_leaves)
    elif node.operator == 'XOR':
        return evaluate(node.left, left_leaves) ^ evaluate(node.right, right_leaves)
    
```

```python
# Adder test 

def run_adder(x,y):
    x = list(reversed(bin(x)[2:]))
    y = list(reversed(bin(y)[2:]))

    n_bits = max(len(x), len(y)) + 1

    values = {} 
    for i in range(n_bits):
        values[f'x{i:02}'] = int(x[i]) if i < len(x) else 0
        values[f'y{i:02}'] = int(y[i]) if i < len(y) else 0
    #print(values)

    result = {}
    for i in range(n_bits):
        result[f'z{i:02}'] = evaluate(nodes[f'z{i:02}'], values)

    return result

def dict_to_int(d):
    return int("".join(map(str, reversed(list(d.values())))), 2)
    
```

```python
# How does an adder work in the first place? 

def bitwise_adder(a, b):
    carry = a & b
    result = a ^ b

    while carry != 0:
        shifted_carry = carry << 1
        carry = result & shifted_carry
        result = result ^ shifted_carry

    return result
```
