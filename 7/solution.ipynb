```python
instructions = []

with open("input.txt", "r") as file:
    for line in file:
        line = line.strip().split()
        test_value = int(line[0][:-1])
        equation_numbers = list(map(int, line[1:]))

        # NOTE: cannot use dict because the test value appears multiple times
        instructions.append((test_value, equation_numbers))
        

instructions
```

```python
total_sum = 0

for test_value, numbers in instructions:

    results = [numbers[0]] 
    

    for number in numbers[1:]: 
        
        temp_results = []
        for result in results:
            temp_results.append(result * number)
            temp_results.append(result + number)
            temp_results.append(int(str(result) + str(number)))

        results = temp_results

    if test_value in results:
        total_sum += test_value
        print(test_value)


# hmm ... this is not working as expected

total_sum
```
